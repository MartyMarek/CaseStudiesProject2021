```python
from pyforest import *
```

```python
repo_rate = pd.read_csv("~/Desktop/RMIT/RMIT SEM 2 2021/Case Studies/Project/Reporoduction_Rate_Data.csv")
variants = pd.read_csv("~/Desktop/RMIT/RMIT SEM 2 2021/Case Studies/Project/Variants_semi_clean.csv")
```

```python
repo_rate["date"]= pd.to_datetime(repo_rate["date"])
variants["date"] = pd.to_datetime(variants["date"])

variant_grp = pd.DataFrame(variants[["date","variant", "num_sequences"]].groupby(["date","variant"]).agg('mean'))
idx = variant_grp.index
variant_grp["date"] = idx.get_level_values(0)
variant_grp["variant"] = idx.get_level_values(1)
variant_grp.index = range(0, len(variant_grp))
```

# Reproduction Rate Graphs

```python
ax = sns.lineplot(x = "date", y = "reproduction_rate", data = repo_rate)
plt.xlabel("Date")
plt.ylabel("Reproduction Rate")
plt.title("Global Reproduction Rate of COVID-19")
plt.xticks(rotation = 90 )
plt.show(ax)
```

```python
ax = sns.lineplot(x = "date", y = "reproduction_rate", hue = "continent", data = repo_rate)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0)
plt.xlabel("Date")
plt.ylabel("Reproduction Rate")
plt.title("Global Reproduction Rate of COVID-19 by continent")
plt.xticks(rotation = 90)
plt.show(ax)

# data needs further grouping to tidy this up. Mean value by date and continent. 
```

```python
ax = sns.relplot(x = "extreme_poverty", y = "reproduction_rate",hue = 'continent', data = repo_rate)
plt.xlabel("Extreme Poverty Value")
plt.ylabel("Reproduction Rate")
plt.title("Rerpoduction Rate of COVID-19 vs Extreme Poverty Rate ")
plt.show(ax)

# need to further refine data for this graph, needs grouping by date and continent and then averaged, but general gist is here
```

# Variants Graphs

```python
ax = sns.lineplot(x = "date", y = "num_sequences", hue = "variant", data = variant_grp)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0, title = "Variant Name" )
plt.xlabel("Date")
plt.ylabel("Number of Sequences")
plt.title("Number of sequences per COVID-19 variants through time")
plt.xticks(rotation = 90)
plt.show(ax)

# need to reduce the number of variants, possibly combine B.1 type variants together or something along those lines
```
