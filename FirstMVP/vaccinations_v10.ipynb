## Vaccinations - Australia

- 151 days of observations of vaccination  data for Australia
- produced using pandas profiling module


```python
import pandas as pd
import pandas_profiling as pp
```

```python
# insert the url if need be.
df = pd.read_csv('./data/owid-covid-data.csv')
df.head()
```

```python
print(df.columns.to_list())
```

```python
# slicing features into groups for EDA. 

# shortened list
cols = ['iso_code', 'continent', 'location', 'date', 'population','total_cases','total_deaths',
        'hospital_beds_per_thousand',  'life_expectancy', 'human_development_index',
        'reproduction_rate', 'total_vaccinations', 'people_vaccinated', 
        'people_fully_vaccinated', 'new_vaccinations', 'new_vaccinations_smoothed', 
        'total_vaccinations_per_hundred', 'people_vaccinated_per_hundred', 
        'people_fully_vaccinated_per_hundred', 'new_vaccinations_smoothed_per_million']

# apply the list 
df = df[cols]

# first pass columns to analyse.  Exploring across 'datasets' if you like.
cols_analyse =  ['total_cases','total_deaths',
        'hospital_beds_per_thousand',  'life_expectancy', 'human_development_index',
        'reproduction_rate', 'total_vaccinations', 'people_vaccinated', 
        'people_fully_vaccinated', 'new_vaccinations', 'new_vaccinations_smoothed', 
        'total_vaccinations_per_hundred', 'people_vaccinated_per_hundred', 
        'people_fully_vaccinated_per_hundred', 'new_vaccinations_smoothed_per_million']

# final.. vaccination columns to analyse.  
vacc_cols_analyse =  ['location','date','total_vaccinations', 'people_vaccinated', 
        'people_fully_vaccinated', 'new_vaccinations', 'new_vaccinations_smoothed', 
        'total_vaccinations_per_hundred', 'people_vaccinated_per_hundred', 
        'people_fully_vaccinated_per_hundred', 'new_vaccinations_smoothed_per_million']
```

### Build vaccination data for AUS.
- filter for loaction = Australia
- susbset the features to only vaccinations
- fill NAs with zeros
- filter for total_vaccinations > 0
- yields 151 obs
- finally reset index to data-location 

```python
df_au = df.query('location == "Australia"')[vacc_cols_analyse].fillna(0).query('total_vaccinations > 0')
df_au.set_index(['date', 'location'], inplace=True, drop=True)
```
